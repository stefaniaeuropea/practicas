Alcance (Reach): El número total de cuentas únicas que han visto una publicación.

```python
alcance_promedio = data_con_relleno['Reach (Organic)'].mean()
alcance_promedio
```

Numero total de acciones realizadas en una publicacion.

```python
interacciones_totales = data_con_relleno['Likes'] + data_con_relleno['Saved'] + data_con_relleno['Comments'] + data_con_relleno['Interactions']
interacciones_promedio = interacciones_totales.mean()
interacciones_promedio
```

tasa de interaccion, la proporcion de interacciones en relacion con el alcance de la publicacion

```python
tasa_interaccion = (interacciones_totales / data_con_relleno['Reach (Organic)']) * 100
tasa_interaccion_promedio = tasa_interaccion.mean()
tasa_interaccion_promedio
```

Tendencias de publicaciones: Analiza la frecuencia de las publicaciones a lo largo del tiempo para identificar patrones y determinar los momentos más efectivos para publicar.

```python
rango_fechas = pd.date_range(start='2020-01-01', end='2023-05-31', freq='M')

# Agrupar los datos por mes y contar el número de publicaciones en cada mes
publicaciones_por_mes = data_con_relleno.groupby(pd.Grouper(key='Timestamp', freq='M')).size()

# Reindexar los datos para incluir todos los meses del rango de fechas
publicaciones_por_mes = publicaciones_por_mes.reindex(rango_fechas, fill_value=0)

# Gráfico de barras para las publicaciones por mes
plt.figure(figsize=(12, 6))
publicaciones_por_mes.plot(kind='bar')
plt.xlabel('Mes')
plt.ylabel('Número de Publicaciones')
plt.title('Tendencia de Publicaciones por Mes')
plt.show()
```

```python
# Filtrar los datos para obtener solo el año 2023
datos_2023 = data_con_relleno.loc[data_con_relleno['Timestamp'].dt.year == 2023]
datos_2023
```

```python
# Agregar una columna para el día de la semana (0: Lunes, 1: Martes, ..., 6: Domingo)
datos_2023['DiaSemana'] = datos_2023['Timestamp'].dt.dayofweek

# Agregar una columna para la hora
datos_2023['Hora'] = datos_2023['Timestamp'].dt.hour
```

```python
# Calcular la frecuencia de publicaciones por día de la semana y hora
frecuencia_publicaciones = datos_2023.groupby(['DiaSemana', 'Hora']).size().unstack(fill_value=0)
```

```python
# Gráfico de mapa de calor para la frecuencia de publicaciones por horas y días de la semana
plt.figure(figsize=(10, 6))
sns.heatmap(frecuencia_publicaciones, cmap='Blues')
plt.xlabel('Hora')
plt.ylabel('Día de la Semana')
plt.title('Tendencia de Publicaciones por Horas y Días de la Semana - 2023')
plt.show()
```
